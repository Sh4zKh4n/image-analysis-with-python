```python
from scipy import misc
from scipy import ndimage as ndi
from skimage import filters
from skimage import feature
```

```python
from skimage import filters
from skimage import feature

def filter_flex(input_image,para,filter_name):
    
    if filter_name=="Median":
        img_out=ndi.median_filter(input_image,para)
    elif filter_name=="Laplace":
        img_out=ndi.gaussian_laplace(input_image,para)
    elif filter_name=="Gaussian":
        img_out=ndi.gaussian_filter(input_image,para)
    elif filter_name=="Sobel":
        img_out=ndi.sobel(input_image,para)
    elif filter_name=="Canny":
        img_out=feature.canny(input_image,para)       
    else:
        print("Coming soon!")
        img_out=input_image
    return (img_out)
```

```python
img1=filter_flex(small,5,"Median")
img2=filter_flex(img1,3,"Gaussian")
img3=filter_flex(img2,2,"Laplace")

    
# define figure
plt.figure(figsize=(7,42)) 
    
# now we plot the results
plt.subplot(6,1,1)
plt.title('Input')
plt.imshow(small,cmap=cm.Greys_r, vmin=0, vmax=small.max())

plt.subplot(6,1,2)
plt.title('After median')
plt.imshow(img1,cmap=cm.Greys_r, vmin=0, vmax=small.max())

plt.subplot(6,1,3)
plt.title('After Gaussian')
plt.imshow(img2,cmap=cm.Greys_r, vmin=0, vmax=small.max())

plt.subplot(6,1,4)
plt.title('After Laplace')
plt.imshow(img3,cmap=cm.Greys_r, vmin=0, vmax=10)


plt.show()
```
